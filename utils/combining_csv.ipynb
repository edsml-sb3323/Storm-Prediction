```python
import pandas as pd

# List all CSV files in the current directory
csv_files = [file for file in os.listdir() if file.endswith('.csv')]

# Initialize an empty DataFrame
combined_df = pd.DataFrame()

# Loop through each CSV file and concatenate to the combined DataFrame
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```

```python
combined_df.to_csv('/content/drive/MyDrive/edsmlproj2/combined_storm.csv', index=False)
```

```python
import pandas

# Read the CSV file into a DataFrame
input_file_path = '/content/drive/MyDrive/edsmlproj2/combined_storm.csv'
df = pd.read_csv(input_file_path)

# Remove '.jpg' from the "image_file_name" column
df['image_file_name'] = df['image_file_name'].str.replace('.jpg', '')

# Rename the column to "image_id"
df = df.rename(columns={'image_file_name': 'image_id'})

# Save the edited DataFrame to a new CSV file
output_file_path = '/content/drive/MyDrive/edsmlproj2/combined_storm_updated.csv'
df.to_csv(output_file_path, index=False)

# Verify the changes
df.head()
```
